```python
weeks = []

for key in dict_id_real:
    weeks.append(key)

start = 564

group_by_week_dict = {}

for week in weeks:
    pairs = []

    for e in dict_id_real[week]:
        
        if e + start in dict_id_fake[week]:
            pairs.append((e, e + start))
        else:
            pairs.append((e, 'DEL'))

    group_by_week_dict[week] = pairs
    print(week, pairs)
    print(len(pairs))
    print()
    start += 1
```

```python

attack_autofill_data_json = {}
for unique_id in unique_ids_list:
    guess_id_in_all_weeks = {}
    for week in group_by_week_dict:
        # get the pair list for one week
        pairs = group_by_week_dict[week]
        # find the unique id in the pair list in this week
        for pair in pairs:
            if pair[0] == unique_id:
                guess_id_in_all_weeks[week] = [str(pair[1])]

        for week in group_by_week_dict:
            if week not in guess_id_in_all_weeks:
                guess_id_in_all_weeks[week] = ["DEL"]

    
    attack_autofill_data_json[unique_id] = guess_id_in_all_weeks

print(attack_autofill_data_json)
```
